#**4. Marco Conceptual (con definiciones básicas)**

**Media:** medida de tendencia central que representa el promedio aritmético de un conjunto de valores numéricos (Elmasri & Navathe, 2015).

**Desviación estándar:** indica cuánto se alejan los datos respecto a la media; mide la dispersión o variabilidad (Montgomery, 2017).

**Outlier:** valor atípico que se desvía significativamente del resto de observaciones y puede afectar las estadísticas descriptivas (Aggarwal, 2015).

**Segmentación:** proceso de dividir los datos en grupos homogéneos para comparar comportamientos o métricas (Han, Kamber & Pei, 2011).

**Visualización de datos:** técnica que transforma información en representaciones gráficas para identificar patrones y tendencias (Few, 2012).

```python
# 11. Transformaciones y limpieza aplicada

# Estado inicial
filas_inicial = df.shape[0]
columnas_inicial = df.shape[1]

# 11.1. Eliminar duplicados exactos
duplicados_inicial = df.duplicated().sum()
df = df.drop_duplicates()

# 11.2. Filtrar registros válidos
qty_invalidos = len(df[df['Quantity'] <= 0])
price_invalidos = len(df[df['UnitPrice'] <= 0])
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# 11.3. Convertir y limpiar tipos de datos
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
fechas_invalidas = df['InvoiceDate'].isnull().sum()

# 11.4. Eliminar registros sin CustomerID
sin_customer = df['CustomerID'].isnull().sum()
df = df.dropna(subset=['CustomerID'])

# 11.5. Normalizar texto
df['Description'] = df['Description'].astype(str).str.strip().str.lower()

# 11.6. Crear columna Total
df['Total'] = df['Quantity'] * df['UnitPrice']

# Resumen de transformaciones
resumen_limpieza = pd.DataFrame({
    'Operación': ['Eliminación duplicados', 'Filtrado Quantity/Price <= 0', 'Conversión fechas', 'Eliminación sin CustomerID', 'Normalización texto', 'Creación columna Total'],
    'Registros_afectados': [duplicados_inicial,qty_invalidos + price_invalidos, fechas_invalidas, sin_customer, 'Todos', 'Todos'],
    'Justificación': ['Evitar conteos duplicados en análisis', 'Solo transacciones válidas para análisis comercial', 'Habilitar análisis temporal', 'Clientes identificables para segmentación', 'Consistencia en búsquedas y agrupamientos', 'Métrica clave para análisis de ingresos']
    })

resumen_limpieza
```

```python
# Impacto final de la limpieza
impacto_final = pd.DataFrame({
    'Métrica': ['Filas iniciales', 'Filas finales', 'Reducción total', 'Columnas iniciales', 'Columnas finales', 'Registros eliminados'],
    'Valor': [filas_inicial, df.shape[0], f"{(1 - df.shape[0]/filas_inicial)*100:.1f}%", columnas_inicial, df.shape[1], filas_inicial - df.shape[0]]
    })

impacto_final
```

```python
# 14. Feature engineering básico

# 14.1. Variables temporales
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['InvoiceYear'] = df['InvoiceDate'].dt.year
df['InvoiceMonth'] = df['InvoiceDate'].dt.month
df['InvoiceDay'] = df['InvoiceDate'].dt.day
df['InvoiceDayOfWeek'] = df['InvoiceDate'].dt.dayofweek
df['InvoiceHour'] = df['InvoiceDate'].dt.hour
df['YearMonth'] = df['InvoiceDate'].dt.to_period('M')

# 14.2. Segmentación por valor de transacción
bins_valor = [0, 10, 50, 100, 500, float('inf')]
labels_valor = ['Muy Bajo', 'Bajo', 'Medio', 'Alto', 'Muy Alto']
df['Segmento_Valor'] = pd.cut(df['Total'], bins=bins_valor, labels=labels_valor)

# 14.3. Categorización de productos por precio
bins_precio = [0, 5, 20, 50, float('inf')]
labels_precio = ['Económico', 'Estándar', 'Premium', 'Lujo']
df['Categoria_Precio'] = pd.cut(df['UnitPrice'], bins=bins_precio, labels=labels_precio)

# Resumen de variables creadas
variables_creadas = pd.DataFrame({
    'Variable': ['InvoiceYear', 'InvoiceMonth', 'InvoiceDay', 'InvoiceDayOfWeek', 'InvoiceHour', 'Segmento_Valor', 'Categoria_Precio', 'YearMonth', 'Total'],
    'Tipo': ['Temporal', 'Temporal', 'Temporal', 'Temporal', 'Temporal', 'Segmentación', 'Categorización', 'Agregación', 'Métrica'],
    'Propósito': ['Análisis por año', 'Análisis mensual', 'Análisis por día', 'Análisis por día semana', 'Análisis por hora', 'Segmentar transacciones por valor', 'Clasificar productos por precio', 'Agrupamiento mensual', 'Ingreso total por transacción']
    })

variables_creadas
```

##**Interpretación de hallazgos**
###1. Distribución geográfica - Concentración Extrema del Mercado
**Hallazgo Principal:** El Reino Unido domina con el **81.6% de los ingresos totales**, confirmando una concentración geográfica extrema.

**Análisis:**
- El mercado doméstico es abrumadoramente dominante, con ingresos 22 veces superiores al segundo mercado (Países Bajos)
- Solo 5 países representan la práctica totalidad del negocio internacional
- **Implicación estratégica:** El modelo de negocio está altamente dependiente del mercado local, presentando tanto fortaleza operativa como vulnerabilidad por falta de diversificación

###2. Estacionalidad mensual - Patrón de Crecimiento con Caída Inesperada
**Hallazgo Principal:** Pico máximo en **noviembre 2011 (€39,116)** y baja sorprendente en **diciembre 2011 (€11,112)**

**Análisis:**
- **Crecimiento consistente:** De febrero a noviembre se observa una tendencia alcista casi ininterrumpida
- **Caída drástica en diciembre:** Contrario a la hipótesis H3, diciembre muestra la peor performance del año
- **Posibles explicaciones:** Cierre contable anticipado, problemas operativos, o cambio estratégico en el período navideño
- **Implicación estratégica:** Revisión de estrategias de inventario y marketing para períodos festivos

###3. Concentración de productos - Distribución Más Equilibrada de lo Esperado
**Hallazgo Principal:** El **top 10 productos representa solo el 9.5%** de los ingresos totales, contradiciendo la hipótesis H1 del principio de Pareto

**Análisis:**
- **Portafolio diversificado:** No existe la concentración extrema esperada (80/20)
- **Producto líder moderado:** "Jumbo bag red retrospot" genera €4,238, representando solo ~1.4% del total
- **Implicación estratégica:** Modelo de negocio basado en volumen amplio más que en productos estrella, requiriendo gestión de inventario más compleja

###4. Comportamiento semanal - Ritmo Comercial Marcado
**Hallazgo Principal:** **Jueves es el día más fuerte (€72,801)** y **Domingo el más débil (€28,908)**

**Análisis:**
- **Variación significativa:** Diferencia del 152% entre el mejor y peor día
- **Patrón previsible:** Jueves-Viernes posiblemente relacionados con pagos quincenales o preparación para fin de semana
- **Fin de semana bajo:** Domingo muestra desempeño consistentemente inferior
- **Implicación estratégica:** Optimización de recursos humanos y logísticos según días de alta demanda

###5. Relación precio-cantidad - Comportamiento de Mercado Saludable
**Hallazgo Principal:** **Correlación negativa de -0.076** confirma la relación inversa esperada entre precio y volumen

**Análisis:**
- **Comportamiento de mercado normal:** Los productos más caros efectivamente se venden en menores cantidades
- **Correlación moderada:** Indica que existen otros factores influyentes además del precio
- **Implicación estratégica:** Estrategias de pricing pueden optimizarse considerando esta relación fundamental
